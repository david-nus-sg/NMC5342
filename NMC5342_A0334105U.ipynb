```python
### PART 3: DATA CLEANING AND KEYWORD ANALYSIS

# --- 1. Install Required Libraries ---
!pip install spacy wordcloud sklearn pandas matplotlib

# Download English spaCy model
!python -m spacy download en_core_web_sm

# --- 2. Import Libraries ---
import re
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
from wordcloud import WordCloud
import spacy
from sklearn.feature_extraction.text import TfidfVectorizer

# --- 3. Load Dataset ---
# Make sure the file is uploaded to /content/
df = pd.read_csv("/content/Instagram-datasets.csv")

# Choose the text column to analyze
text_column = "comment"  # <-- change this if needed
texts = df[text_column].dropna().astype(str).tolist()

# --- 4. Initialize spaCy Model ---
nlp = spacy.load("en_core_web_sm")

# --- 5. Cleaning Function ---
def clean_text(text):
    # Remove URLs
    text = re.sub(r"http\S+|www\S+", "", text)
    # Remove emojis and special characters
    text = re.sub(r"[^\w\s]", "", text)
    # Remove extra spaces
    text = re.sub(r"\s+", " ", text).strip()
    return text

# --- 6. Tokenization, Lemmatization, Stopword Removal ---
def preprocess_text_spacy(text):
    doc = nlp(text.lower())  # lowercase
    tokens = [
        token.lemma_  # lemmatize
        for token in doc
        if not token.is_stop and token.is_alpha  # remove stopwords & non-alphabetic
    ]
    return tokens

# --- 7. Clean + Preprocess All Texts ---
cleaned_texts = [clean_text(t) for t in texts]
tokenized_texts = [preprocess_text_spacy(t) for t in cleaned_texts]

# --- 8. Generate Word Frequency Table ---
all_tokens = [token for sublist in tokenized_texts for token in sublist]
word_freq = Counter(all_tokens)
freq_df = pd.DataFrame(word_freq.most_common(20), columns=["word", "frequency"])

print("ğŸ”  Top 20 Most Frequent Words:")
display(freq_df)

# --- 9. Generate Word Cloud ---
wordcloud = WordCloud(
    width=800,
    height=400,
    background_color='white'
).generate(" ".join(all_tokens))

plt.figure(figsize=(10,5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.title("Word Cloud of Cleaned Text", fontsize=16)
plt.show()

# --- 10. Calculate TFâ€“IDF Scores ---
vectorizer = TfidfVectorizer(tokenizer=lambda x: x, lowercase=False)
tfidf_matrix = vectorizer.fit_transform(tokenized_texts)
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=vectorizer.get_feature_names_out())

print("ğŸ“Š TFâ€“IDF Scores (first 5 rows):")
display(tfidf_df.head())

# --- 11. Save Results ---
freq_df.to_csv("/content/word_frequency.csv", index=False)
tfidf_df.to_csv("/content/tfidf_scores.csv", index=False)
print("âœ… Files saved: word_frequency.csv and tfidf_scores.csv")

###Based on the code run, the top recurring words are de, love, e, que, beautiful
```

```python
### PART 4: SENTIMENT AND TOPIC MODELLING

## Task 1: Sentiment Analysis

# ============================================
# ğŸ’¬ SENTIMENT ANALYSIS PIPELINE (Colab Version)
# ============================================

# --- 1. Install Required Libraries ---
!pip install textblob matplotlib pandas nltk

# --- 2. Import Libraries ---
import pandas as pd
import matplotlib.pyplot as plt
from textblob import TextBlob
import nltk

# Download tokenizer used by TextBlob
nltk.download('punkt')

# --- 3. Load Dataset ---
# Make sure your file is uploaded to /content/
df = pd.read_csv("/content/Instagram-datasets.csv")

# Set the text column (change this if your column has a different name)
text_column = "comment"   # e.g., "content", "text", etc.
df[text_column] = df[text_column].astype(str)

# --- 4. Compute Polarity & Subjectivity ---
def analyze_sentiment(text):
    blob = TextBlob(text)
    return blob.sentiment.polarity, blob.sentiment.subjectivity

df[["polarity", "subjectivity"]] = df[text_column].apply(
    lambda x: pd.Series(analyze_sentiment(x))
)

# --- 5. Assign Sentiment Labels ---
def label_sentiment(score):
    if score > 0.1:
        return "Positive"
    elif score < -0.1:
        return "Negative"
    else:
        return "Neutral"

df["sentiment_label"] = df["polarity"].apply(label_sentiment)

# --- 6. Display Summary Statistics ---
print("Average Polarity:", round(df["polarity"].mean(), 3))
print("Average Subjectivity:", round(df["subjectivity"].mean(), 3))
print("\nSentiment Distribution:")
print(df["sentiment_label"].value_counts())

# --- 7. Visualization: Bar Chart of Sentiment Categories ---
plt.figure(figsize=(6,4))
df["sentiment_label"].value_counts(normalize=True).mul(100).plot(
    kind="bar",
    color=["green", "gray", "red"]
)
plt.title("Sentiment Distribution (%)")
plt.ylabel("Percentage of Posts")
plt.xticks(rotation=0)
plt.show()

# --- 8. Visualization: Histogram of Polarity Scores ---
plt.figure(figsize=(8,5))
plt.hist(df["polarity"], bins=20, edgecolor="black")
plt.title("Distribution of Polarity Scores")
plt.xlabel("Polarity (-1 = Negative, +1 = Positive)")
plt.ylabel("Number of Posts")
plt.show()

# --- 9. Save Results to CSV ---
df.to_csv("/content/sentiment_results.csv", index=False)
print("âœ… Sentiment results saved as /content/sentiment_results.csv")

## Task 2 Topic Modelling

# --- 1. Install Dependencies ---
!pip install gensim pyLDAvis spacy pandas matplotlib
!python -m spacy download en_core_web_sm

# --- 2. Import Libraries ---
import pandas as pd
import gensim
from gensim import corpora
from gensim.models import LdaModel
import pyLDAvis.gensim
import spacy
import matplotlib.pyplot as plt

# --- 3. Load Dataset ---
df = pd.read_csv("/content/Instagram-datasets.csv")

# Choose text column (change if needed)
text_column = "comment" # Changed from "caption" to "comment"
texts = df[text_column].dropna().astype(str).tolist()

# --- 4. Initialize spaCy Model ---
nlp = spacy.load("en_core_web_sm")

# --- 5. Text Cleaning and Tokenization ---
import re

def clean_text(text):
    text = re.sub(r"http\S+|www\S+", "", text)   # remove URLs
    text = re.sub(r"[^\w\s]", "", text)          # remove emojis/symbols
    text = re.sub(r"\s+", " ", text).strip()
    return text

def preprocess_spacy(text):
    doc = nlp(text.lower())
    return [token.lemma_ for token in doc if not token.is_stop and token.is_alpha]

cleaned_texts = [clean_text(t) for t in texts]
tokenized_texts = [preprocess_spacy(t) for t in cleaned_texts]

# --- 6. Create Dictionary and Corpus ---
dictionary = corpora.Dictionary(tokenized_texts)
corpus = [dictionary.doc2bow(text) for text in tokenized_texts]

print(f"Dictionary size: {len(dictionary)} unique tokens")
print(f"Corpus size: {len(corpus)} documents")

# --- 7. Train LDA Model (Try 3â€“5 Topics) ---
num_topics = 5  # change to 3 if you prefer fewer topics
lda_model = LdaModel(
    corpus=corpus,
    id2word=dictionary,
    num_topics=num_topics,
    random_state=42,
    passes=10,
    alpha='auto',
    per_word_topics=True
)

# --- 8. Display Keywords per Topic ---
print("\nğŸ§© LDA Topics and Top Words:\n")
topics = lda_model.print_topics(num_words=10)
for idx, topic in topics:
    print(f"Topic {idx + 1}: {topic}\n")

# --- 9. Visualize Topics (Interactive) ---
import pyLDAvis
pyLDAvis.enable_notebook()
lda_display = pyLDAvis.gensim.prepare(lda_model, corpus, dictionary)
pyLDAvis.display(lda_display)

# --- 10. Optional: Save Visualization ---
pyLDAvis.save_html(lda_display, "/content/lda_topics.html")
print("âœ… Interactive LDA visualization saved as lda_topics.html")

# --- 11. Brief Topic Interpretation ---
print("\nğŸ§  Topic Descriptions (Based on Top Words):\n")

for idx, topic in topics:
    words = [w.split("*")[1].replace('"', '').strip() for w in topic.split("+")]
    summary = ", ".join(words[:6])
    print(f"Topic {idx + 1}: Likely relates to themes around {summary}.")
```

```python
### Part 5: Network Analysis

## Task 1: User Interaction Analysis

# --- 1. Install Required Libraries ---
!pip install pandas networkx matplotlib pyvis

# --- 2. Import Libraries ---
import pandas as pd
import re
import networkx as nx
import matplotlib.pyplot as plt
from pyvis.network import Network

# --- 3. Load Dataset ---
df = pd.read_csv("/content/Instagram-datasets.csv")

# ğŸ”§ Change these to match your dataset's column names:
user_column = "comment_user"           # the user who posted
caption_column = "comment"         # text that may contain mentions like @someone

# --- 4. Extract Mentions from Text ---
def extract_mentions(text):
    # Finds all words starting with '@'
    mentions = re.findall(r"@(\w+)", str(text))
    return mentions

df["mentions"] = df[caption_column].apply(extract_mentions)

# --- 5. Build User Interaction Edges ---
edges = []
for _, row in df.iterrows():
    user = row[user_column]
    for mentioned_user in row["mentions"]:
        if user and mentioned_user:
            edges.append((user, mentioned_user))

print(f"Extracted {len(edges)} user-to-user interactions.")

# --- 6. Create and Analyze Network ---
G = nx.DiGraph()
G.add_edges_from(edges)

print(f"Network contains {len(G.nodes())} users and {len(G.edges())} interactions.")

# --- 7. Calculate Centrality Metrics (optional insights) ---
degree_centrality = nx.degree_centrality(G)
top_users = sorted(degree_centrality.items(), key=lambda x: x[1], reverse=True)[:10]

print("\nğŸ” Top 10 Most Connected Users:")
for user, score in top_users:
    print(f"{user}: {round(score, 3)}")

# --- 8. Static Visualization (Matplotlib) ---
plt.figure(figsize=(10, 8))
pos = nx.spring_layout(G, k=0.3)
nx.draw(
    G,
    pos,
    with_labels=True,
    node_size=50,
    font_size=8,
    arrowstyle='->',
    arrowsize=10
)
plt.title("User Interaction Network")
plt.show()

# --- 9. Interactive Visualization (PyVis) ---
net = Network(height="750px", width="100%", bgcolor="#222222", font_color="white")

# Add nodes and edges
for node in G.nodes():
    net.add_node(node, title=node)

for edge in G.edges():
    net.add_edge(edge[0], edge[1])

# Generate interactive visualization
net.show("/content/user_interaction_network.html")
print("âœ… Interactive network visualization saved as /content/user_interaction_network.html")
```

```python
## Task 4: Compute Network Metrics

# --- 1. Install Required Libraries ---
!pip install pandas networkx matplotlib pyvis

# --- 2. Import Libraries ---
import pandas as pd
import re
import networkx as nx
import matplotlib.pyplot as plt
from pyvis.network import Network
from itertools import combinations

# --- 3. Load Dataset ---
df = pd.read_csv("/content/Instagram-datasets.csv")

# Columns
user_column = "comment_user"
text_column = "comment"

# Ensure text is string
df[text_column] = df[text_column].astype(str)

# ===============================
# 4A. USER INTERACTION NETWORK
# ===============================

# Extract mentions
def extract_mentions(text):
    return re.findall(r"@(\w+)", text)

df["mentions"] = df[text_column].apply(extract_mentions)

# Build edge list
user_edges = df[[user_column, "mentions"]].explode("mentions")
user_edges = user_edges.rename(columns={user_column: "source", "mentions": "target"})
user_edges = user_edges.dropna(subset=["target"]).reset_index(drop=True)

# Create NetworkX graph
G_user = nx.DiGraph()
G_user.add_edges_from(user_edges[["source", "target"]].values)

print(f"User Network: {len(G_user.nodes())} nodes, {len(G_user.edges())} edges")

# ===============================
# 4B. HASHTAG CO-OCCURRENCE NETWORK
# ===============================

# Extract hashtags
def extract_hashtags(text):
    return re.findall(r"#(\w+)", text.lower())

df["hashtags"] = df[text_column].apply(extract_hashtags)

# Build hashtag co-occurrence edges
hashtag_edges = []
for hashtags in df["hashtags"]:
    # Consider all pairs in the same post
    if len(hashtags) > 1:
        hashtag_edges.extend(combinations(hashtags, 2))

# Create NetworkX graph
G_hashtag = nx.Graph()
G_hashtag.add_edges_from(hashtag_edges)

print(f"Hashtag Network: {len(G_hashtag.nodes())} nodes, {len(G_hashtag.edges())} edges")

# ===============================
# 5. Visualize Networks
# ===============================

# --- 5A. Static Visualization (Users) ---
plt.figure(figsize=(10,8))
pos = nx.spring_layout(G_user, k=0.3)
nx.draw(G_user, pos, with_labels=False, node_size=50, node_color="skyblue", edge_color="gray", alpha=0.6)
plt.title("User Interaction Network")
plt.show()

# --- 5B. Static Visualization (Hashtags) ---
plt.figure(figsize=(10,8))
pos = nx.spring_layout(G_hashtag, k=0.3)
nx.draw(G_hashtag, pos, with_labels=True, node_size=50, node_color="orange", edge_color="gray", alpha=0.6, font_size=8)
plt.title("Hashtag Co-occurrence Network")
plt.show()

# --- 5C. Interactive Visualization (Users) ---
net_user = Network(height="600px", width="100%", directed=True, bgcolor="#222222", font_color="white")
net_user.from_nx(G_user)
net_user.show_buttons(filter_=['physics'])
net_user.show("/content/user_network.html")

# --- 5D. Interactive Visualization (Hashtags) ---
net_hashtag = Network(height="600px", width="100%", directed=False, bgcolor="#222222", font_color="white")
net_hashtag.from_nx(G_hashtag)
net_hashtag.show_buttons(filter_=['physics'])
net_hashtag.show("/content/hashtag_network.html")

print("âœ… Interactive networks saved as user_network.html and hashtag_network.html")
```
